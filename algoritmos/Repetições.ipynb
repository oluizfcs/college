```python
#11 Ler números até que 0 seja inserido, mostre a quantidade e a média de ímpares/pares e a média geral.
quantidade_impares = 0
quantidade_pares = 0
quantidade_total = 0
soma_impares = 0
soma_pares = 0
soma = 0

print("Insira números inteiros positivos e \"0\" quando quiser parar")

while True:

  numero = int(input(f"Número {quantidade_total + 1}: "))

  if numero == 0:
    break

  quantidade_total += 1
  soma += numero

  if numero % 2 == 0:
    quantidade_pares += 1
    soma_pares += numero
  else:
    quantidade_impares += 1
    soma_impares += numero

print(f"\n Ímpares: {quantidade_impares} Média: {soma_impares / quantidade_impares}")
print(f"\n Pares: {quantidade_pares} Média: {soma_pares / quantidade_pares}")
print(f"\n Total: {quantidade_total} Média: {soma / quantidade_total}")
```

```python
#14 análise da população
contador = 0
soma_salario = 0
soma_filhos = 0
baixa_renda = 0
maior_salario = 0
while True:

  print(f"\n Família {contador + 1}:")

  salario = float(input("\n Salário da família: "))
  if salario < 0:
    break

  if salario <= 100:
    baixa_renda += 1

  if salario > maior_salario:
    maior_salario = salario

  soma_salario += salario

  filhos = int(input("\n Quantidade de filhos: "))
  soma_filhos += filhos

  contador += 1

print(f"Média salarial da população (por família) -> R$: {soma_salario / contador}")
print(f"Média de filhos por família: {soma_filhos / contador}")
print(f"Maior salário: {maior_salario}")
print(f"Percentual de pessoas com salário até R$: 100,00 -> {(baixa_renda / contador) * 100}%")
```

```python
#28 n-ésimo número de Fibonacci onde n é informado pelo usuário
n = int(input("n: "))
contador = 2

termo1 = 1
termo2 = 1
termo_atual = 0

if n == 1 or n == 2:
  print(1)
else:

  while contador < n:
    termo_atual = termo1 + termo2
    termo1 = termo2
    termo2 = termo_atual

    contador += 1

  print(termo_atual)
```

```python
x = 2
perfeitos = 0

while perfeitos < 6:

  divisores = 0
  contador = 0

  while contador <= x or divisores < 2:

    contador += 1
    resto = x % contador

    if resto == 0:
      divisores += 1

  if divisores <= 2:
    # O número é primo
    print(2 ** (x - 1) * ((2 ** x) - 1))
    perfeitos += 1

  x += 1
```

```python
x = 2
perfeitos = 0

while perfeitos < 8:

  divisores = 0
  contador = 0

  while contador <= x or divisores < 2:

    contador += 1
    resto = x % contador

    if resto == 0:
      divisores += 1

  if divisores <= 2:
    # O número é primo
    possivel_perfeito = (2 ** (x - 1) * ((2 ** x) - 1))

    soma_divisores = 0

    for y in range(1, possivel_perfeito):
      if possivel_perfeito % y == 0:
        soma_divisores += y

    if soma_divisores == possivel_perfeito:
      perfeitos += 1
      print(f"{perfeitos}º = {possivel_perfeito}")

  x += 1
```
