```python
pergunta = input("Você sabe o raio? (sim/não) ");

def valor(pergunta):
  if pergunta == "sim":
    raio = int(input("Digite o raio do círculo: "))
  elif pergunta == "não":
    raio = (int(input("Digite o diâmetro do círculo: ")))/2
  return raio

raio = valor(pergunta)
PI = 3.14159265 # constante
area = (raio * raio) * PI

print("Área do círculo de raio", raio, "é:", round(area, 2))
```

```python
def imparOuPar(n):
  if n % 2 == 0:
    print('É par')
  else:
    print('É ímpar')

imparOuPar(358)
```
