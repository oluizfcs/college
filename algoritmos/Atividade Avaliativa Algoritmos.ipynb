Escrever um algoritmo que leia o nome de um vendedor, o seu salário fixo e o total de vendas efetuadas por ele no mês (em dinheiro). Sabendo que este vendedor ganha 15% de comissão sobre suas vendas efetuadas, informar o seu nome, o salário fixo e salário no final do mês.



Observação: Sempre suponha a leitura de dados do teclado e a exibição de saídas no monitor. Ainda, construa programas elegantes, que apresentam mensagens de requisição de dados ao usuário.

```python
#Questão 3
nome_do_vendedor = input("Nome do vendedor: ")
salario_fixo = float(input("Salário do vendedor: "))
vendas = float(input("Valor arrecadado em vendas: "))

comissao = vendas * 0.15
salario_final = salario_fixo + comissao

print(f"\n Nome: {nome_do_vendedor} \n Salário fixo: R$ {salario_fixo:,.2f} \n Salário final do mês: R$ {salario_final:,.2f}")
```

Crie um algoritmo para ler uma temperatura em graus Celsius e apresentá-la convertida em graus Fahrenheit. A fórmula de conversão é: F=(9*C+160) / 5, sendo F a temperatura em Fahrenheit e C a temperatura em Celsius.



Observação: Sempre suponha a leitura de dados do teclado e a exibição de saídas no monitor. Ainda, construa programas elegantes, que apresentam mensagens de requisição de dados ao usuário.

Faça um algoritmo para ler três notas de um aluno em uma disciplina e imprimir a sua média ponderada (as notas têm pesos respectivos de 1, 2 e 3).



Observação: Sempre suponha a leitura de dados do teclado e a exibição de saídas no monitor. Ainda, construa programas elegantes, que apresentam mensagens de requisição de dados ao usuário.

```python
#Questão 5
notas = input("Insira as três notas de um aluno separadas por vírgula (Ex: 5.9, 6, 7.1): ")
notas = list(map(float, notas.split(",")))

nota_1 = notas[0] * 1
nota_2 = notas[1] * 2
nota_3 = notas[2] * 3

media = (nota_1 + nota_2 + nota_3) / 6

print(f"Média ponderada do aluno: {media:.2f}")
```

Escreva um programa para calcular e mostrar o salário semanal de uma pessoa, determinado pelas condições que seguem. Se o número de horas trabalhadas for inferior a 40, a pessoa recebe R$15,00 por hora, senão a pessoa recebe R$600,00 mais R$21,00 para cada hora trabalhada acima de 40 horas. O programa deve pedir o número de horas trabalhadas como entrada e deve dar o salário como saída.

```python
#Questão 8
horas_trabalhadas = int(input("Horas trabalhadas: "))

salario = 0

if horas_trabalhadas < 40:
  salario = horas_trabalhadas * 15
else:
  salario = (horas_trabalhadas - 40) * 21 + 600

print(f"Salário semanal = R$: {salario:,.2f}")
```

Escreva um algoritmo que leia três valores inteiros e verifique se eles podem ser os lados de um triângulo. Se forem, informar qual o tipo de triângulo que eles formam: equilátero, isósceles ou escaleno.

Propriedades:

* O comprimento de cada lado de um triângulo é menor do que a soma dos comprimentos dos outros dois lados;
* Triângulo Equilátero: aquele que tem os comprimentos dos três lados iguais;
* Triângulo Isósceles: aquele que tem o comprimento de dois lados iguais;
* Triângulo Escaleno: aquele que tem os comprimentos de seus três lados diferentes

```python
#Questão 10
lados = input("Insira 3 valores inteiros separados por vírgula (Ex: 2, 3, 4): ")
lados = list(map(int, lados.split(",")))
lados.sort(reverse = True)

A = lados[0]
B = lados[1]
C = lados[2]

if A >= B + C:
  print("Os valores inseridos não podem ser lados de um triângulo.")
else:
  if A != B and B != C and C != A:
    print("Triângulo escaleno")
  if A == B and B == C:
    print("Triângulo equilátero")
  elif A == B or B == C:
    print("Triângulo isósceles")
```
