```python
from datetime import datetime

formato = '%d/%m/%Y %H:%M:%S'

data_inicio = datetime.strptime('13/04/2006 12:42:00', formato)
data_atual = datetime.strptime('13/04/2024 12:42:00', formato)

delta = data_atual - data_inicio

print(delta.days, delta.total_seconds())
```

```python
from datetime import datetime, timedelta

formato = '%d/%m/%Y %H:%M:%S'
data = datetime.strptime('01/01/2024 00:00:00', formato)
delta = timedelta(days=1, hours=-3)


print(data + delta)
```
