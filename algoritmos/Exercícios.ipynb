```python
# 02
n1 = int(input("Insira o primeiro número  "))
n2 = int(input("Insira o segundo número  "))

soma = n1 + n2
sub = n1 - n2
multi = n1 * n2
div = n1 / n2

r_soma = f"{n1} + {n2} = {soma}"
r_sub = f"{n1} - {n2} = {sub}"
r_multi = f"{n1} * {n2} = {multi}"
r_div = f"{n1} / {n2} = {div}"

print(f"Resultados: \n {r_soma} \n {r_sub} \n {r_multi} \n {r_div}")
```

```python
# 03
valor = float(input("Insira o valor do produto: "))
pago = float(input("Insira o valor pago:  "))

troco = pago - valor

if valor > pago:
  print("O valor pago é insuficiente, ficou faltando R$: ", troco)
elif valor == pago:
  print("Não tem troco")
elif valor < pago:
  print("O troco é de R$:", troco)
```

```python
# 13
deposito = int(input("Qual o valor depositado?  "))
meses = int(input("Quantos meses esse dinheiro vai ficar rendendo?  "))

JURO = 0.7 / 100

rendimento = deposito + deposito * JURO

i = 1
while i <= meses:
  print(f"mês {i} = R$:{round(rendimento, 2)}")
  rendimento = rendimento + rendimento * JURO
  i += 1
```

```python
#16
numeros = input("Insira os coeficientes 'a', 'b', 'c', 'd', 'e' e 'f' separados por vírgula. ")

n = list(map(int, numeros.split(",")))

x = (n[2] * n[4] - n[1] * n[5]) / n[0] * n[4] - n[1] * n[3]
y = (n[0] * n[5] - n[2] * n[3]) / n[0] * n[4] - n[1] * n[3]

print(f"x = {x}, y = {y}")
```

```python
#17
hora = int(input("Insira o número de horas trabalhadas: "))
extra = int(input("Insira o número de horas extras trabalhadas: "))

hora_normal = 10
hora_extra = 15

bruto = (hora * hora_normal) + (extra * hora_extra)

liquido = bruto - bruto * 0.1

print(f"\n Salário bruto = R$:{bruto} \n Salário líquido = R$:{liquido}")
```

```python
#21
carga = int(input("Carga horária da disciplina: "))
horas_aula = int(input("Horas/aula dessa disciplina: "))
faltas = int(input("Faltas: "))

aulas = carga / horas_aula

presente = aulas - faltas

frequencia = (presente * 100) / aulas

print(f"Percentual de faltas = {100 - frequencia}%")
```
