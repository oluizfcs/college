### Exercícios em Sala

```python
lista1 = [1, 2, 3]
lista2 = [2, 3, 4]

lista3 = lista1 + lista2

i = 0

while i < len(lista3):

  if i in lista1 and i in lista2:
    lista3.remove(i)

  i += 1

print(lista3)
```

```python
lista_de_inteiros = [3, 4, 27, 1, 6, 2, 2, 10]
pares = []
impares = []

for x in lista_de_inteiros:
  if x % 2 == 0:
    pares.append(x)
  else:
    impares.append(x)

print(f"Pares = {pares}")
print(f"Ímpares = {impares}")
```

```python
lista_de_inteiros = [5, -25, 72, 26 -5238, 1, 0, 10]
maior_numero = 0

for x in lista_de_inteiros:
  if x > maior_numero:
    maior_numero = x

print(maior_numero)
```
