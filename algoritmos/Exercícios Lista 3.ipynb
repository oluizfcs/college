```python
#6
ano_atual = int(input("Ano atual: "))
ano_nascimento = int(input("Ano de nascimento: "))

idade = ano_atual - ano_nascimento

def calcularIdade(idade):

  if idade >= 0 and idade <= 3:
    x = "bebê"
  elif idade >= 4 and idade <= 10:
    x = "criança"
  elif idade >= 11 and idade <= 18:
    x = "adolescente"
  elif idade >= 19 and idade <= 60:
    x = "adulta"
  elif idade > 60:
    x = "idosa"
  print(f"Essa pessoa é um(a) {x} pois tem {idade} anos.")

if idade < 1:
    print(f"Essa pessoa ainda não nasceu, pois teria {idade} anos.")
else:
  calcularIdade(idade)
```

```python
#8
nome = input("Nome: ")
idade = int(input("Idade: "))
grupo = input("Grupo de risco (Baixo, Médio ou Alto): ").lower()
categoria = 0

if grupo == "baixo" or grupo == "médio" or grupo == "alto":

  if idade >= 17 and idade <= 20:
    if grupo == "baixo":
      categoria = 1
    elif grupo == "médio":
      categoria = 2
    elif grupo == "alto":
      categoria = 3

  elif idade >= 21 and idade <= 24:
    if grupo == "baixo":
      categoria = 2
    elif grupo == "médio":
      categoria = 3
    elif grupo == "alto":
      categoria = 4

  elif idade >= 25 and idade <= 34:
    if grupo == "baixo":
      categoria = 3
    elif grupo == "médio":
      categoria = 4
    elif grupo == "alto":
      categoria = 5

  elif idade >= 35 and idade <= 64:
    if grupo == "baixo":
      categoria = 4
    elif grupo == "médio":
      categoria = 5
    elif grupo == "alto":
      categoria = 6

  elif idade >= 65 and idade <= 70:
    if grupo == "baixo":
      categoria = 7
    elif grupo == "médio":
      categoria = 8
    elif grupo == "alto":
      categoria = 9

  else:
    print("A idade deve ser entre 17 e 70 anos.")

  print(f"\n {nome} como você tem {idade} anos e está no grupo de risco {grupo}, a sua categoria é {categoria}")
else:
  print("Grupo inválido")
```

```python
#10
nome = input("Nome: ")
sexo = input("Sexo Masculino/Feminino:")
altura = float(input("Altura: "))
idade = int(input("Idade: "))

if sexo.lower() == "masculino":

  if altura > 1.7:

    if idade <= 20:
      print(f"Peso ideal = {(72.7 * altura) - 58}kg")
    elif idade >= 21 and idade <= 39:
      print(f"Peso ideal = {(72.7 * altura) - 53}kg")
    elif idade >= 40:
      print(f"Peso ideal = {(72.7 * altura) - 45}kg")

  elif altura <= 1.7:

    if idade <= 40:
      print(f"Peso ideal = {(72.7 * altura) - 50}kg")
    elif idade > 40:
      print(f"Peso ideal = {(72.7 * altura) - 58}kg")

elif sexo.lower() == "feminino":

  if altura > 1.5:

    if idade <= 20:
      print(f"Peso ideal = {(62.1 * altura) - 44.7}kg")

  elif altura <= 1.5:

    if idade >= 35:
      print(f"Peso ideal = {(62.1 * altura) - 45}kg")
    elif idade < 35:
      print(f"Peso ideal = {(62.1 * altura) - 49}kg")

else:
  print("Sexo Inválido")
```
